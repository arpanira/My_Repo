```python

import requests
import pandas as pd
import json
import os
import fnmatch
pd.set_option('display.max_colwidth',100)
```

## Text Acquistion using API Call:

```python
#Intialise file path and filename where API call response  will be stored .

filepath='C:/Users/Home/Datasceince_Nov18/POC_Work/movie_json/'
filename='output_movie_file*.json'

#There are total 457 pages, but we will only use fetch few pages
total_pages=30

# Let us remove any old files present in the directory.
#Check if any file matches the filename pattern
files_in_directory = os.listdir(filepath)
matching_files = fnmatch.filter(files_in_directory, filename)
print(matching_files)
if matching_files:
    for file_name in matching_files:
        file_path = os.path.join(filepath, file_name)
        os.remove(file_path)
        print(f"The file {file_name} has been removed.")
else:
     print("Files not present")
        
        
#Loop to fetch response from 30 pages and Write content into a seperate json file

for i in range(1,total_pages+1):
  # Specify the API endpoint URL
  url = f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}'
  response = requests.get(url)
  # Check if the request was successful (status code 200)
  if response.status_code == 200:
        
        data = response.json()
        file_name = f'{filepath}/output_movie_file_{i}.json'
        with open(file_name,"w") as file:
            json.dump(data,file)
        df=pd.read_json(file_name)
        dfs.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
## Concatenate along rows (axis=0) example: "result = pd.concat([df1, df2])""


final_df = pd.concat(dfs, ignore_index=True)

# Display the final DataFrame

print(final_df.head(5))
```
